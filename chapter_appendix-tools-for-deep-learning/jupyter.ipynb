## 8-1 مدولاسیون فرکانس باریک‌باند

مدولاسیون فرکانس با سیگنال زیر را در نظر بگیرید:

$$
x(t) = A \cos(\omega_m t) \tag{8-8}
$$

توجه به معادلات (8-34)، (8-37)، فرکانس لحظه‌ای برابرست با:

$$
\omega_i(t) = \omega_c + k_f A \cos(\omega_m t) \tag{8-39}
$$

که به صورت سینوسی بین مقادیر
$$
\omega_c - k_f A
$$
و
$$
\omega_c + k_f A
$$
تغییر می‌کند. با تعریف:

$$
\Delta \omega = k_f A
$$

داریم:

$$
\omega_i(t) = \omega_c + \Delta \omega \cos(\omega_m t)
$$

و

$$
y(t) = \cos\left( \omega_c t + \int x(t) \, dt \right)
$$

که برابر می‌شود با:

$$
y(t) = \cos\left( \omega_c t + \frac{\Delta \omega}{\omega_m} \sin(\omega_m t) + \theta_0 \right) \tag{8-40}
$$

که در آن
$$
\theta_0
$$
ثابت انتگرال‌گیری است. برای سهولت فرض می‌کنیم
$$
\theta_0 = 0
$$
. پس:

$$
y(t) = \cos\left( \omega_c t + \frac{\Delta \omega}{\omega_m} \sin(\omega_m t) \right) \tag{8-41}
$$

عامل
$$
\Delta \omega / \omega_m
$$
را **شاخص مدولاسیون فرکانس** تعریف می‌کنیم. سیستم‌های FM بسته به بزرگ یا کوچک بودن شاخص مدولاسیون
$$
m = \Delta \omega / \omega_m
$$
ویژگی‌های متفاوتی دارند. حالتی که
$$
m
$$
کوچک است FM باریک‌باند می‌باشد. به طور کلی می‌توان معادله‌ی (8-41) را به صورت زیر نوشت:

$$
y(t) = \cos\left( \omega_c t + m \sin(\omega_m t) \right) \tag{8-42}
$$

یا

$$
y(t) = \cos(\omega_c t) \cos(m \sin(\omega_m t)) - \sin(\omega_c t) \sin(m \sin(\omega_m t)) \tag{8-43}
$$

اگر
$$
m
$$
بسیار کوچک باشد (خیلی کوچکتر از
$$
\pi/2
$$
)، می‌توان تقریب‌های زیر را استفاده کرد:

$$
\cos(m \sin(\omega_m t)) \approx 1
$$

$$
\sin(m \sin(\omega_m t)) \approx m \sin(\omega_m t)
$$

(معادله‌ی 8-44)

بنابراین معادله‌ی (8-43) چنین می‌شود:

$$
y(t) = \cos(\omega_c t) - m \sin(\omega_m t) \sin(\omega_c t)
$$

طبق (1) (براساس تقریب در شکل (8-33)) نشان داده شده است. توجه شود که این طیف با طیف AM-DSB/WC شباهت دارد. شباهت‌ها عبارتند از: از نظر این لحاظ که دو فرکانس حامل در طرف موجود است و کنار باندهای وجود دارد که در مدولاسیون AM مشابه دیده می‌شود.

در حالی که تفاوت‌های مدلوله کننده همزمان است و در معادله‌ی (8-46) موج FM باریک‌باند می‌بینیم که سیگنال حاصل با حاصل ضرب مدوله شده‌اند. در حالی که در AM-DSB/WC فقط حاصل جمع طیف‌ها وجود دارد.

شکل 8-34 شکل موج FM باریک‌باند مربوط به معادله‌ی (8-46) را نشان می‌دهد.

</div>


```python
import numpy as np
import matplotlib.pyplot as plt

# تعریف مکان پیک‌ها
omega_c = 2  # مکان ωc
delta = np.pi / 2  # فاصله mπ/2
main_peak = 1.0    # ارتفاع پیک مرکزی
side_peak = 0.6    # ارتفاع پیک‌های جانبی

# موقعیت فرکانسی فلش‌ها
frequencies = [-omega_c - delta, -omega_c, -omega_c + delta, 0, omega_c - delta, omega_c, omega_c + delta]

# جهت فلش‌ها (بالا یا پایین)
directions = [-1, 1, 1, 1, -1, 1, 1]  # منفی یعنی فلش به سمت پایین

# اندازه هر فلش
amplitudes = [side_peak, main_peak, side_peak, main_peak, side_peak, main_peak, side_peak]

# رسم
plt.figure(figsize=(12, 5))
plt.hlines(0, -omega_c - delta - 1, omega_c + delta + 1, color='black')  # محور افقی

# رسم فلش‌ها
for x, y, d in zip(frequencies, amplitudes, directions):
    plt.arrow(x, 0, 0, d * y, head_width=0.1, head_length=0.08, fc='black', ec='black')

# برچسب گذاری دقیق
plt.text(-omega_c, -0.2, r'$-\omega_c$', ha='center', fontsize=12)
plt.text(omega_c, -0.2, r'$\omega_c$', ha='center', fontsize=12)
plt.text(0, -0.2, r'$0$', ha='center', fontsize=12)

plt.text(-omega_c - delta, -side_peak - 0.25, r'$\frac{m\pi}{2}$', ha='center', fontsize=12)
plt.text(-omega_c + delta, side_peak + 0.05, r'$-\frac{m\pi}{2}$', ha='center', fontsize=12)
plt.text(omega_c - delta, -side_peak - 0.25, r'$-\frac{m\pi}{2}$', ha='center', fontsize=12)
plt.text(omega_c + delta, side_peak + 0.05, r'$\frac{m\pi}{2}$', ha='center', fontsize=12)

# تنظیمات گراف
plt.xlabel(r'$\omega$', fontsize=14)
plt.ylabel(r'$Y(j\omega)$', fontsize=14)
plt.title('Spectrum with Up and Down Arrows (Correct)')
plt.xlim(-omega_c - delta - 1, omega_c + delta + 1)
plt.ylim(-1.4, 1.4)
plt.grid(False)
plt.show()
```

## شکل 8-34 مقایسه FM باند باریک و AM-DSB/WC

(الف) FM باند باریک  
(ب) AM-DSB/WC

</div>


```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
fc = 20  # فرکانس حامل (Hz)
fm = 2   # فرکانس پیام (Hz)
m = 0.5  # ضریب مدولاسیون کوچک برای باریک‌باند

# محور زمان
t = np.linspace(0, 2, 1000)  # 2 ثانیه

# موج FM باریک‌باند
y_fm = np.cos(2*np.pi*fc*t) - m * np.sin(2*np.pi*fm*t) * np.sin(2*np.pi*fc*t)

# موج AM-DSB/WC
y_am = np.cos(2*np.pi*fc*t) + m * np.cos(2*np.pi*fm*t) * np.cos(2*np.pi*fc*t)

# رسم شکل‌ها
plt.figure(figsize=(14, 8))

# شکل الف: FM باند باریک
plt.subplot(2, 1, 1)
plt.plot(t, y_fm, color='black')
plt.title('shekl (alef) - FM کیراب دناب')
plt.xlabel('t (socond)')
plt.ylabel('y(t)')
plt.grid(True)

# شکل ب: AM-DSB/WC
plt.subplot(2, 1, 2)
plt.plot(t, y_am, color='black')
plt.title('شکل (ب) - AM-DSB/WC')
plt.xlabel('t (second)')
plt.ylabel('y(t)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

$$
m \sin(\omega_m t)
$$
مدوله شده است.

\omega_c

\omega_c \pm n \omega_m

$$
\omega_c
$$
و قرار گرفتن فرکانس‌های
$$
\omega_c \pm \omega_m, \omega_c \pm 2\omega_m, \ldots
$$
تشکیل داده است.

معادله (8-8) از ضرایب سری فوریه در ناحیه ضرایب سری فوریه انتقال پیدا می‌کند و بنابراین طیف معادله (8-8) محدود به طیف پایه
$$
\omega_c

$$
n
$$
از
$$
\omega_m

## 8-7-3 سیگنال مدوله کننده مربعی متناوب

مثال دیگری که به درک ویژگی‌های مدولاسیون فرکانس کمک می‌کند، سیگنال مدوله کننده موج مربعی است.  
در معادله (8-39) قرار دهید:

$$
A = 1.5, \quad f_m = 5Hz, \quad \omega_m = 10\pi
$$

و فرض کنید که شکل 8-36 فرم سیگنال مدوله شده شده است.  
(در شکل 8-37) نشان داده شده است. به ازای $X(t)$ مثبت، فرکانس لحظه‌ای
$$
\omega_c + \Delta \omega
$$
و به ازای $X(t)$ منفی،
$$
\omega_c - \Delta \omega
$$
می‌شود. بنابراین می‌توان به صورت زیر نوشت:

$$
y(t) = r(t) \cos\left[(\omega_c + \Delta \omega)t\right] + [1-r(t)] \cos\left[(\omega_c - \Delta \omega)t\right] \tag{8-50}
$$

که در آن $r(t)$ موج مربعی متناوب شکل 8-36 است.  
بنابراین برای این سیگنال مدوله کننده خاص، تعیین طیف سیگنال
$$
y(t)
$$
معادل تعیین طیف مجموع دو سیگنال AM مطابق معادله (8-50) است. دقیق‌تر این که:

$$
Y(j\omega) = \frac{1}{\gamma} \left[ R(j\omega + j\omega_c + j\Delta\omega) + R(j\omega - j\omega_c + j\Delta\omega) \right]
+ \frac{1}{\gamma} \left[ Rr(j\omega + j\omega_c - j\Delta\omega) + Rr(j\omega - j\omega_c - j\Delta\omega) \right] \tag{8-51}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
A = 1            # دامنه
T = 2            # تناوب موج مربعی
fc = 10          # فرکانس حامل ωc (Hz)
delta_f = 2      # انحراف فرکانسی Δf (Hz)
fm = 1 / T       # فرکانس موج مربعی
t = np.linspace(-T, T, 4000)  # محور زمان (بازه زیادتر برای شکل تمیزتر)

# تعریف موج مربعی r(t)
r = np.zeros_like(t)
for i in range(len(t)):
    time_mod = np.mod(t[i] + T/2, T)
    if (0 <= time_mod < T/4) or (3*T/4 <= time_mod < T):
        r[i] = 0  # بخش منفی
    else:
        r[i] = 1  # بخش مثبت

# تعریف y(t)
y = A * (r * np.cos(2*np.pi*(fc + delta_f)*t) + (1 - r) * np.cos(2*np.pi*(fc - delta_f)*t))

# رسم موج
plt.figure(figsize=(12, 4))
plt.plot(t, y, color='black')

# محور و توضیحات
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.text(0, 1.2, r'$y(t)$', ha='center', fontsize=12)
plt.xlabel('Time (t)', fontsize=14)
plt.ylabel('Amplitude', fontsize=14)
plt.title('Figure 8-37: Frequency Modulation with Square Wave Modulator')
plt.xlim(-T, T)
plt.ylim(-1.5, 1.5)
plt.grid(True)
plt.show()
```

### ۸-۸-۱ مدولاسیون دامنه سینوسی گسسته در زمان

$$

## شکل ۸-۳۹

اندازه‌ی طیف مدولاسیون فرکانس با سیگنال مدوله کننده موج مربعی.  
نشان می‌دهد که طیف در نقاط
$\omega_c$،
$\omega_c \pm \Delta\omega$،
$\omega_c \pm 3\Delta\omega$،
$\omega_c \pm 5\Delta\omega$
و غیره خطوط فرکانسی دارد و ارتفاع هر خط متناسب با ضریب فوریه‌ی متناظر است.

</div>


```python
import numpy as np
import matplotlib.pyplot as plt

# پارامترها
omega_c = 0        # مرکز ωc
delta_omega = 1    # فاصله خطوط اصلی (Δω)
num_lines = 30     # تعداد خطوط هر طرف
shift = 40         # شیفت طیف دوم

# موقعیت خطوط
frequencies = []
heights = []

for n in range(-num_lines, num_lines + 1):
    freq = omega_c + n * delta_omega
    frequencies.append(freq)
    height = np.exp(-0.15 * abs(n))  # کاهش نمایی
    heights.append(height)

# رسم
plt.figure(figsize=(16, 5))
plt.hlines(0, -10, 70, color='gray')  # محور ω

# دسته اول (مرکز ωc)
for f, h in zip(frequencies, heights):
    plt.arrow(f, 0, 0, h, head_width=0.3, head_length=0.02, fc='black', ec='black')

# دسته دوم (مرکز ωc + shift)
for f, h in zip(frequencies, heights):
    plt.arrow(f + shift, 0, 0, h, head_width=0.3, head_length=0.02, fc='black', ec='black')

# برچسب گذاری فقط برای ωc و چند نقطه مهم
label_positions = [0, -1, 1, 3, shift, shift - 1, shift + 1, shift + 3]
label_texts = [
    r'$\omega_c$', r'$(\omega_c-\Delta\omega)$', r'$(\omega_c+\Delta\omega)$', r'$(\omega_c+3\Delta\omega)$',
    r'$\omega_c$', r'$(\omega_c-\Delta\omega)$', r'$(\omega_c+\Delta\omega)$', r'$(\omega_c+3\Delta\omega)$'
]

for pos, text in zip(label_positions, label_texts):
    plt.text(pos, -0.15, text, ha='center', fontsize=12)

# تنظیمات گراف
plt.xlabel(r'$\omega$', fontsize=14)
plt.ylabel(r'$Y(j\omega)$', fontsize=14)
plt.title('Figure 8-39 (Two Repeated Spectra - Neatly Arranged)')
plt.xlim(-10, 70)
plt.ylim(-0.2, 1.2)
plt.grid(False)
plt.show()
```
